```python
import torch
import torch.nn as nn
from network import net

m = 0

D = net.AuxDiscriminator_128(hidden_dim=512, projection_dim=128)
D_target = net.TargetNetwork_128(hidden_dim=512, projection_dim=128)

D_param = D.named_parameters()
D_target = D_target.named_parameters()

D_target_param_dict = dict(D_target)

print('Target')
print(D_target_param_dict['res_blocks.0.block.0.bias'])

for name, param in D_param:
    if name == 'res_blocks.0.block.0.bias':
        print('Online')
        print(param.data)

#     # Copy
#     if name in D_target_param_dict:
#         D_target_param_dict[name].data.copy_(param.data)
        
    # Momentum update
    # if m == 0 -> copy
    if name in D_target_param_dict:
        online_param = param.data
        target_param = D_target_param_dict[name].data
        new_target_param = target_param * m + online_param * (1. - m)
        D_target_param_dict[name].data.copy_(new_target_param)
        
print('New target')
print(D_target_param_dict['res_blocks.0.block.0.bias'])
```
